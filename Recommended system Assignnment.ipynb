```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler


# Convert 'episodes' to numeric (handle 'object' type)
data['episodes'] = pd.to_numeric(data['episodes'], errors='coerce')
data['episodes'].fillna(data['episodes'].mean(), inplace=True) 

# Normalize 'rating' and 'members' columns
scaler = MinMaxScaler()
data[['rating', 'members']] = scaler.fit_transform(data[['rating', 'members']])

# Use TF-IDF for genres (convert to a numerical representation)
tfidf = TfidfVectorizer(stop_words='english')
genre_matrix = tfidf.fit_transform(data['genre'])

# Encode the 'type' column (e.g., 'TV', 'Movie', 'OVA')
label_encoder = LabelEncoder()
data['type_encoded'] = label_encoder.fit_transform(data['type'])

# Combine features into a single dataframe for similarity calculation
features = pd.concat([
    pd.DataFrame(genre_matrix.toarray(), columns=tfidf.get_feature_names_out()),
    data[['rating', 'members', 'type_encoded', 'episodes']]
], axis=1)
features
```

# Recommendation System:

Design a function to recommend anime based on cosine similarity.

Given a target anime, recommend a list of similar anime based on cosine similarity scores.

Experiment with different threshold values for similarity scores to adjust the recommendation list size.


```python
# Check for missing values
features.isnull().sum()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
cos_sim = cosine_similarity(features)

def recommend_anime(target_anime_id, top_n=10):
    # Get index of the target anime
    target_idx = data[data['anime_id'] == target_anime_id].index[0]
    
    
    similarity_scores = cos_sim[target_idx]
    
    # Get indices of the top N similar animes (excluding the target anime itself)
    similar_anime_idx = similarity_scores.argsort()[-top_n-1:-1][::-1]
    
   
    recommended_animes = data.iloc[similar_anime_idx][['name', 'anime_id']]
    recommended_animes['similarity_score'] = similarity_scores[similar_anime_idx]
    
    return recommended_animes

# Example usage: recommend top 5 animes similar to 'Kimi no Na wa.'
recommend_anime(32281, top_n=5)

```

# Evaluation:

Split the dataset into training and testing sets.


```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

```

Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score.

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_recommendations(target_anime_id, top_n=10):
    # Get top N recommendations for the target anime
    recommended_animes = recommend_anime(target_anime_id, top_n)
    
    # Get the list of anime ids recommended
    recommended_anime_ids = recommended_animes['anime_id'].tolist()
    
    
    is_target_in_recommendations = 1 if target_anime_id in recommended_anime_ids else 0
    
    
    ground_truth = 1 if target_anime_id in test_data['anime_id'].values else 0
    
    return is_target_in_recommendations, ground_truth
```

```python
# Evaluate the system
y_true = []
y_pred = []

# Loop over all test data anime
for target_anime_id in test_data['anime_id']:
    is_target_in_recommendations, ground_truth = evaluate_recommendations(target_anime_id, top_n=10)
    y_true.append(ground_truth)
    y_pred.append(is_target_in_recommendations)

# Calculate precision, recall, and F1-score
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

Analyze the performance of the recommendation system and identify areas of improvement.

```python
# Loop to evaluate the system with different top_n values
for top_n in [10, 20, 50]:
    y_true = []
    y_pred = []
    
    # Loop through all test data anime
    for target_anime_id in test_data['anime_id']:
        is_target_in_recommendations, ground_truth = evaluate_recommendations(target_anime_id, top_n)
        y_true.append(ground_truth)
        y_pred.append(is_target_in_recommendations)
    
    # Calculate precision, recall, and F1-score for the current top_n
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    # Print the results
    print(f"Top-N: {top_n}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
```

# 1. Difference Between User-Based and Item-Based Collaborative Filtering:

User-Based Collaborative Filtering: Recommends items based on the preferences of similar users. If User A likes items that User B likes, items liked by User B will be recommended to User A.

Item-Based Collaborative Filtering: Recommends items similar to the ones the user has already interacted with. If User A liked "Item X", the system recommends similar items to "Item X" based on other users' preferences.

Key Difference:

User-based focuses on user similarity, while item-based focuses on item similarity.

# 2. What is Collaborative Filtering and How Does It Work?

Collaborative Filtering is a recommendation system technique that suggests items based on the behavior of similar users or items. It works in two main ways:

Memory-Based: Uses user-item interaction data to calculate similarities (e.g., user-based or item-based).

Model-Based: Uses machine learning (e.g., matrix factorization) to generate recommendations based on hidden patterns in the data.

How it works:

Collect user interaction data (ratings, clicks, etc.).

Calculate similarity between users or items.

Recommend items based on these similarities.

Advantages:

Does not require item details or user demographics.

Learns and adapts based on user behavior.

Disadvantages:

Cold start problem (new users/items have no data).

Can be computationally expensive.
