# Task 3: Clustering with Original Data:

```python
# normalising the data
norm_data=MinMaxScaler().fit_transform(data)
norm_data
```

# 1 fitting into the kmeans clustering model 

The Elbow Method is a useful technique for identifying the right number of clusters

so below the plot we can take cluster are 3



```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.cluster import KMeans
WCSS=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i).fit(norm_data)
    WCSS.append(kmeans.inertia_)
plt.plot(range(1,11),WCSS)
plt.title("The elbow method")
plt.xlabel("no of clusters")
plt.ylabel("WCSS")
plt.show()
    
```

```python
kmodel=KMeans(n_clusters=3).fit(norm_data)
kmodel
```

```python
kmodel.inertia_ # gives the sum of square means  are also called within cluster sum of sqaures
```

```python
kmodel.labels_ # this appends the cluster value to each data
```

```python
kmodel.cluster_centers_ # gives the dimensions  belonging  to certain clustes
```

```python
# converting the labels to data frame to adding the orginal data 
labels_data=pd.DataFrame(kmodel.labels_,columns=['clusters'])
labels_data
```

# 2.Visualize the clustering results using appropriate plots.

```python
plt.figure(figsize=(10,5))
plt.scatter(norm_data[:,12],norm_data[:,13],c=kmodel.labels_,s=50,cmap='viridis')
# ploting the centriods
centroids = kmodel.cluster_centers_
plt.scatter(centroids[:, 12], centroids[:, 13], c='red', s=200, alpha=0.50, marker='X', label='Centroids')

plt.title('K-Means Clustering Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

# 3.Evaluate the clustering performance using metrics such as silhouette score 

silhouette measures quality of clusters

Closer to 1: Indicates well-separated and cohesive clusters.

Closer to -1: Indicates poor clustering with points potentially assigned to the wrong cluster.

Around 0: Indicates overlapping clusters

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(norm_data,kmodel.labels_)
score # output shows around 0 so indicates overlapping clusters
```

# 4.Analyze the trade-offs between using PCA and clustering directly on the original dataset.

Dimensionality Reduction: PCA compresses the data, reducing the number of features used for clustering. This helps improve the performance but might lose some small amount of information that could have been relevant for understanding certain nuances in the original data.

Faster Computation: Clustering in a reduced-dimensional space (after PCA) is computationally faster and more efficient, especially for high-dimensional datasets, compared to clustering in the full original feature space.
