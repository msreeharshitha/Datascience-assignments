```python
#data visuliazation
#histigram
import seaborn as sns
import matplotlib.pyplot as plt
for i in df.select_dtypes(include="number").columns:
    sns.histplot(data=df,x=i)
    plt.show()
```

```python
#finding outliers and interquartile range
des=numerical_columns.describe()
des
```

```python
for i in des:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    IQR=q3-q1
    upperlimit=q3+1.5*IQR
    lowerlimit=q3-1.5*IQR
    print(f"i:{i}")
    print(f"Q1 (25th percentile): {q1}")
    print(f"Q3 (75th percentile): {q3}")
    print(f"IQR(interquartile range):{IQR}")
    print(f"upperlimit:{upperlimit}")
    print(f"lowerlimit:{lowerlimit}")
    print("-" * 40)
```

```python
#printing outliers
outliers = df[(df[i] < lowerlimit) | (df[i] > upperlimit)]

outliers
```

```python
#printing data after removing outliers
no_outliers=df[(df[i]>=lowerlimit)&(df[i]<=upperlimit)]
no_outliers
```

```python
for i in no_outliers.select_dtypes(include="number").columns:
    sns.boxplot(data=no_outliers,x=i)
    plt.show()
```

```python
# summary after removing outliers
des=no_outliers.describe()
des
```

```python
# Bar Chart Analysis for Categorical Column:
for i in df.select_dtypes(include="object").columns:
    categori_counts=df[i].value_counts()
    sns.barplot(x=categori_counts.index,y=categori_counts.values,palette='viridis')
    plt.title(f"bar chart of:{i}")
    plt.xlabel(i)
    plt.ylabel("frequency")
    plt.xticks(rotation=90)
    plt.show()
```
