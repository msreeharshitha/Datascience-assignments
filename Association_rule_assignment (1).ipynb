 # Data Preprocessing

```python
dummies_df=pd.get_dummies(data).astype(int)
dummies_df
```

# Association Rule Mining:

Apply the Apriori Algorithm

The apriori function generates frequent itemsets that have a minimum support level.

```python
frequent_itemsets= apriori(dummies_df,min_support=0.005,use_colnames=True)
frequent_itemsets
```

```python
print(association_rules.__module__)
print(frequent_itemsets.head())
print(frequent_itemsets.dtypes)

```

```python
num_itemsets = frequent_itemsets.shape[0] 
num_itemsets
```

Set appropriate threshold for support, confidence and lift to extract meaning full rules.
this gives antecedents and consequents with lift metrics

```python
rules = association_rules(frequent_itemsets, 
                          num_itemsets=num_itemsets, 
                          metric="lift", 
                          min_threshold=0)
rules
```

```python
rules[rules.lift>1]
```

so here  all have lift metric is > 1 so we can say there are positively related


# Interview Questions:
# 1.What is lift and why is it important in Association rule

lift : measures the strength  of  association rule 

calculated by Lift(A → B)= Confidence(A → B)/Support(B)
                                                           
lift=1:A and B are independent

lift>1:A and B are positively related mean occuring more together than expected 

lift<1:A and B are negatively related mean occuring less together than expected 

Importance:

Lift helps identify meaningful and actionable associations.

which have strong lift value can be prioritize first


# 2.What is support and Confidence. How do you calculate them?

Support:

Indicates how frequently an itemset appears in the dataset.

Formula:support(A)= Transactions(A)/total transactions 

Purpose: Filters out infrequent items, ensuring rules are based on significant patterns.

Confidence:

Measures the likelihood that item B is purchased when item A is purchased.

Formula:Confidence(A → B)=Support(A ∩ B)/support A

Purpose: Evaluates the reliability of a rule. Higher confidence indicates stronger predictability.



# What are some limitations or challenges of Association rules mining?
High Dimensionality:
Large datasets with many items lead to an exponential increase in potential rules, making it computationally expensive.
Support-Confidence 

Tradeoff:
Rules with high confidence might have low support, making them less useful for practical applications.

Redundancy:
Many rules may overlap or convey similar information, complicating analysis.

Interpretability:
Too many rules can overwhelm users, making it hard to derive actionable insights.
