# Discuss the pros and cons of One-Hot Encoding and Label Encoding.
# One-Hot Encoding

Pros:

Avoids introducing ordinal relationships (ideal for nominal variables).

Works well for non-ordinal data.

Can improve model accuracy, especially for algorithms like decision trees.

Widely applicable across many machine learning algorithms.

Cons:

Increases the feature space, especially with high-cardinality variables.

Leads to sparse matrices (many zeros), which can be inefficient.

May cause overfitting when there are too many categories.


# Label Encoding

Pros:

Simple and efficient (doesnâ€™t increase the number of features).

Memory efficient.

Works well for ordinal data (where categories have a meaningful order).

Cons:

Imposes an ordinal relationship that may not exist in nominal data.

Can mislead some algorithms into thinking there's a mathematical relationship between categories.

Not suitable for nominal data, where no ordering is implied.

How Outliers Affect Model Performance

1. Skewed Metrics:
Regression Models: Outliers can significantly distort metrics like Mean Squared Error (MSE), as these metrics are sensitive to extreme values.

2. Classification Models: Outliers may mislead the decision boundary, reducing accuracy and generalizability.

3. Overfitting:
Outliers might force the model to overfit to these rare points, leading to poor performance on unseen data.

4. Feature Distribution Distortion:
Outliers can affect scaling and normalization, making features less interpretable and harder to use effectively.


5. Increased Variance:
Models might exhibit high variance due to the influence of outliers, making predictions unstable.

```python
import ppscore as pps
pps_matrix = pps.matrix(data)
pps_matrix
```

```python
correlation_matrix = data.corr()
correlation_matrix
```

```python
# Create a heatmap for PPS
plt.figure(figsize=(17,9))
pps_heatmap = pps_matrix.pivot(columns="x", index="y", values="ppscore")
sns.heatmap(pps_heatmap, annot=True, cmap="Blues")
plt.title("PPS Matrix Heatmap")
plt.show()
```

```python
plt.figure(figsize=(17,9))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

The **correlation matrix** identifies strong linear relationships, such as `hours_per_week` and `hours_per_year`, but misses non-linear or categorical 

dependencies. In contrast, the **Predictive Power Score (PPS)** detects non-linear and categorical relationships, revealing that features like

`education_num` and `income_ >50K` or `sex_ Female` and `anomaly` may have strong predictive power, even when their correlation is low or undefined.

PPS is more versatile for feature importance in prediction tasks, while correlation is best for linear relationships in numerical data.
