#  Split the dataset into training and testing sets (80% training, 20% testing).

```python
x=data.iloc[:,0:16]
y=data.iloc[:,16]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(f'x_train: {x_train}')
print('----' * 10)

print(f'x_test: {x_test}')
print('----' * 10)

print(f'y_train: {y_train}')
print('----' * 10)

print(f'y_test: {y_test}')
print('----' * 10)
```

#  Implement the K-Nearest Neighbours algorithm using a machine learning library like scikit-learn On training dataset

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
accuracies=[]
k_values=range(1,20)
for i in k_values:
    model=KNeighborsClassifier(n_neighbors=i).fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracies.append(accuracy_score(y_test,y_pred))
                      
print(accuracies)
```

```python
import matplotlib.pyplot as plt

plt.plot(k_values, accuracies, marker='o', linestyle='-', color='b')
plt.title('Accuracy vs k for KNN')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

#  Choose an appropriate distance metric and value for K.

# so by seeing the plot we can say that k=5 is best but choosing the best k value based on good accuracy rises to overfiting and underfiting

i.e Small k: When k is small (e.g., 1 or 3) is data have some noise and leads to overfiting

Large 𝑘: When k is large, the model becomes smoother and may underfit


```python
best_knn=KNeighborsClassifier(n_neighbors=5).fit(x_train,y_train)
best_knn
```

```python
y_pred=best_knn.predict(x_test)
y_pred
```

# Evaluate the classifier's performance on the testing set using accuracy, precision, recall, and F1-score metrics.

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
accuracy=accuracy_score(y_test,y_pred)
precision = precision_score(y_test, y_pred, average='micro')  # 'micro' for multiclass
recall = recall_score(y_test, y_pred, average='micro')  # 'micro' for multiclass
f1 = f1_score(y_test, y_pred, average='micro')
print(f'accuracy:{accuracy}')
print(f'precision:{precision}')
print(f'recall:{recall}')
print(f'f1:{f1}')
```

# the other way is using cross validation which avoides overfiting

cross valadiation is technique the divides the data set into k equal sub sets and perform the model for each set and gives a avergae accuracy score of them

```python
import warnings 
warnings.filterwarnings("ignore")

from sklearn.model_selection import cross_val_score
model2=KNeighborsClassifier(n_neighbors=3)
k_fold=cross_val_score(model2,x_train,y_train,cv=5)
print(k_fold)
```

# . Visualize the decision boundaries of the classifier.

```python
from sklearn.decomposition import PCA
# Reduce dimensions for visualization
pca = PCA(n_components=2)
x_train_2d = pca.fit_transform(x_train)
x_test_2d = pca.transform(x_test)

# Train the KNN model
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust n_neighbors
knn.fit(x_train_2d, y_train)

# Evaluate the model
y_pred = knn.predict(x_test_2d)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Plot decision boundaries
x_min, x_max = x_train_2d[:, 0].min() - 1, x_train_2d[:, 0].max() + 1
y_min, y_max = x_train_2d[:, 1].min() - 1, x_train_2d[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
plt.scatter(x_train_2d[:, 0], x_train_2d[:, 1], c=y_train, edgecolor='k', marker='o', cmap=plt.cm.Paired, label='Train')
plt.scatter(x_test_2d[:, 0], x_test_2d[:, 1], c=y_test, edgecolor='k', marker='x', cmap=plt.cm.Paired, label='Test')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KNN Decision Boundaries')
plt.legend()
plt.show()
```

# What are the key hyperparameters in KNN?

   
Number of Neighbors (n_neighbors)

Definition: The number of nearest neighbors the algorithm considers when making a prediction.

Distance Metric (metric)

Definition: The method used to calculate the distance between data points.

 Weighting Function (weights)
 
Definition: Determines how the neighbors contribute to the prediction.

Tuning KNN Hyperparameters

To find the optimal hyperparameter values, you can use techniques like:

Grid Search: Systematically search through a set of hyperparameter combinations.

Random Search: Randomly sample combinations of hyperparameters.

Cross-Validation: Evaluate model performance on different folds of data to avoid overfitting.



#  What distance metrics can be used in KNN?

Distance Metric (metric)

Definition: The method used to calculate the distance between data points.

Common Choices:

Euclidean: Default; works well with continuous data.

Manhattan: Suitable for grid-like data or when features have different scales.

Minkowski: Generalization of Euclidean and Manhattan; allows adjusting the power parameter p.

Custom Metric: You can define your own distance function.

Effect: Choosing the wrong metric can degrade performance.
