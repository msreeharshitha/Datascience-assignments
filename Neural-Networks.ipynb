#  Model Implementation

dividing the data into train and test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_df,y_df,test_size=0.3,random_state=42)
print(f'x_train:{x_train}')
print('----'*10)
print(f'x_test:{x_test}')
print('----'*10)
print(f'y_train:{y_train}')
print('----'*10)
print(f'y_test:{y_test}')

```

A basic ANN with one hidden layer  

```python
import tensorflow as tf
from tensorflow import keras
model1=keras.Sequential([
     keras.layers.Dense(64, input_shape=(x_train.shape[1],), activation='relu'),  # Hidden layer
    keras.layers.Dense(26, activation='softmax')  # Output layer
    ])
model1.compile(
    optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy']
)
model1.fit(x_train,y_train,epochs=5)
```

Train your model on the training set and then use it to make predictions on the test set.

```python
y_pred=model1.predict(x_test)
y_pred
```

# Hyperparameter Tuning

●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.

```python
import tensorflow as tf
from tensorflow import keras
model2=keras.Sequential([
    keras.layers.Dense(128 ,input_shape=(x_train.shape[1],),activation='relu'),#input layer
    keras.layers.Dense(64,activation='relu'),# hidden layer
    keras.layers.Dense(26,activation='softmax')# output layer
])

model2.compile(
    optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy']
)
model2.fit(x_train,y_train,epochs=50)
```

```python
pred=model2.predict(x_test)
pred
```

Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.

i have used random search

```python
from keras_tuner import RandomSearch
def build_model(hp):
    model = keras.Sequential([
        keras.layers.Dense(hp.Choice('units', [32, 64, 128]), activation=hp.Choice('activation', ['relu', 'tanh'])),
        keras.layers.Dense(26, activation='softmax')
    ])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Configure and run the tuner
tuner = RandomSearch(build_model, objective='val_accuracy', max_trials=5, directory='.', project_name='alphabet_classification')
tuner.search(x_train, y_train, epochs=5, validation_split=0.2)

# Get and print the best hyperparameters
best_hps = tuner.get_best_hyperparameters(1)[0]
print(f"Best params: Units={best_hps.get('units')}, Activation={best_hps.get('activation')}, LR={best_hps.get('learning_rate')}")

# Train and evaluate the best model
best_model = tuner.hypermodel.build(best_hps)
best_model.fit(x_train, y_train, epochs=10, validation_split=0.2)
print(f"Test Accuracy: {best_model.evaluate(x_test, y_test)[1]:.4f}")
```

```python
y_pred2 = best_model.predict(x_test)
y_pred2
```
