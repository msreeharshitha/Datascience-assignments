# : SVM Implementation

I have used PCA to improve model performance and to visulaize the SVM results

```python
from sklearn.decomposition import PCA
xtrain_pca=PCA(n_components=2).fit_transform(x_train)
xtrain_pca
```

```python
xtest_pca=PCA(n_components=2).fit_transform(x_test)
xtest_pca
```

so I use class weighting method which can adjust the weights and helps model in focusing on minority class

```python
from sklearn.svm import SVC
model=SVC(class_weight='balanced',kernel='linear').fit(xtrain_pca,y_train)
model
```

```python
y_pred=model.predict(xtest_pca)
y_pred 
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
accuracy=accuracy_score(y_test,y_pred)
precision=accuracy_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)

accuracy
```

# Visualization of SVM Results

1.	Visualize classification results on the testing data.


```python

# Step 1: Create a meshgrid for the decision boundary
x_min, x_max = xtrain_pca[:, 0].min() - 1, xtrain_pca[:, 0].max() + 1
y_min, y_max = xtrain_pca[:, 1].min() - 1, xtrain_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Step 2: Predict for meshgrid points (Decision Boundary)
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 3: Plot the results
plt.figure(figsize=(10, 6))

# Plot the decision boundary
plt.contourf(xx, yy, Z, levels=50, cmap='coolwarm', alpha=0.8)  # Decision boundary
plt.contour(xx, yy, Z, levels=[-1, 0, 1], colors=['blue', 'black', 'red'], linestyles=['--', '-', '--'])

# Scatter plot of testing data
plt.scatter(xtest_pca[:, 0], xtest_pca[:, 1], c=y_test, cmap='plasma', edgecolors='k', label='Testing Data')

# Highlight support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            facecolors='none', edgecolors='k', s=100, label='Support Vectors')

# Customize plot
plt.title('SVM Decision Boundary with RBF Kernel on Testing Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid()
plt.show()

```

# Parameter Tuning and Optimization

# parameter tunning

Common Hyperparameters to Tune

Support Vector Machine (SVM):

C: Regularization parameter that controls the trade-off between achieving a low error on the training data and minimizing model complexity (overfitting).

kernel: Defines the type of kernel to be used (linear, RBF, polynomial, etc.).

gamma: Kernel coefficient for non-linear kernels (affects the shape of the decision boundary).

degree: Degree of the polynomial kernel (used only for the poly kernel).

# Comparison and Analysis

1.Compare SVM performance with various kernels (e.g., linear, polynomial, radial basis function).
  
Linear Kernel: Performs reasonably well with balanced precision and recall for both classes, but with lower overall accuracy (46%).

RBF Kernel: Achieves high accuracy (67%) but is biased toward class 1, with perfect recall for class 1 and failure to predict class 0 (0% precision and recall for class 0).

Polynomial Kernel: Likely to fall between the performance of the linear and RBF kernels, offering a balance but requiring testing to confirm.

In conclusion, the Linear Kernel provides a more balanced approach, while the RBF Kernel offers better performance for class 1 but fails with class 0.

2.	Analyze SVM strengths and weaknesses for the Mushroom dataset based on EDA and visualization results.

Strengths of SVM for Mushroom Dataset:

Handles High Dimensionality: SVM works well with both categorical and continuous features (e.g., cap_shape, odor).

Clear Margin of Separation: Effective for datasets with distinct class boundaries (e.g., edible vs. poisonous mushrooms).

Non-Linear Boundaries: SVMs can model complex decision boundaries with kernels.

Suitable for Small Datasets: The Mushroom dataset is small, making SVM a good fit.

Weaknesses of SVM for Mushroom Dataset:

Computational Complexity: SVMs can be slow and memory-intensive with larger datasets.

Hyperparameter Sensitivity: Requires careful tuning of parameters like the kernel and C to avoid overfitting.

Lack of Interpretability: SVMs, especially with non-linear kernels, are harder to interpret.

Preprocessing Needs: Categorical data (e.g., cap_shape, odor) requires encoding before use.

Overall, SVM is effective for the Mushroom dataset but requires careful tuning and preprocessing.
